```python
class bst:
  def __init__(self,key):
    self.key = key
    self.lchild = None
    self.rchild = None

  def insert(self,data):
    if self.key is None:
      self.key = data
      return

    elif self.key == data:
      return

    elif self.key > data:
      if self.lchild:
        self.lchild.insert(data)
      else:
        self.lchild = bst(data)

    else:
      if self.rchild:
        self.rchild.insert(data)
      else:
        self.rchild = bst(data)

  def search(self,data):
    if self.key == data:
      print('Node is found, you mongrol!')
      return

    elif self.key < data:
      if self.rchild:
        self.rchild.search(data)
      else:
        print('Node is not present in the tree you mongrol!')

    else:
      if self.lchild:
        self.lchild.search(data)
      else:
        print('Node is not present in the tree you mongrol!')

  def preorder(self): # root -> left -> right
    print(self.key, end = " ")
    if self.lchild:
      self.lchild.preorder()
    if self.rchild:
      self.rchild.preorder()

  def inorder(self): # left -> root -> right
    if self.lchild:
      self.lchild.inorder()
    print(self.key, end = " ")
    if self.rchild:
      self.rchild.inorder()

  def postorder(self): # left -> right -> root
    if self.lchild:
      self.lchild.postorder()
    if self.rchild:
      self.rchild.postorder()
    print(self.key, end = " ")

  def delete(self, data, current):
    if self.key is None:
      print("Tree is empty you mongrol. Can't delete anything")
    if data < self.key:
      if self.lchild:
        self.lchild = self.lchild.delete(data, current)
      else:
        print("Given Node is not present in the tree you mongrol")

    elif data > self.key:
      if self.rchild:
        self.rchild = self.rchild.delete(data, current)
      else:
        print("Given Node is not present in the tree you mongrol")

    else:    # Conditions for the root node
      if self.lchild is None:
        temp = self.rchild
        if data == current:
          self.lchild = temp.lchild
          self.rchild = temp.rchild
          temp = None
          return temp
        self = None
        return temp

      if self.rchild is None:
        temp = self.lchild
        if data == current:
          self.lchild = temp.lchild
          self.rchild = temp.rchild
          temp = None
          return temp
        self = None
        return temp
      
      node = self.rchild
      while node.lchild:
        node = node.lchild
      self.key = node.key
      self.rchild = self.rchild.delete(node.key, current)
    
    return self

  def min_node(self):
    current = self
    while current.lchild:
      current = current.lchild
    print('Node with smallest key is',current.key)


  def max_node(self):
    current = self
    while current.rchild:
      current = current.rchild
    print('Node with largest key is',current.key)

      




root = bst(10)
list1 = [20,11,6,13,3,8,12,15,1,100]
for i in list1:
  root.insert(i)


print('You want to traverse the tree in Preorder way, fine, then have it your way you mongrol:')
root.preorder()



print('\nYou want to traverse the tree in Inorder way, fine, then have it your way you mongrol:')
root.inorder()

print('\nYou want to traverse the tree in Postorder way, fine, then have it your way you mongrol:')
root.postorder()


def count(node):
  if node is None:
    return 0
  return 1 + count(node.lchild) + count(node.rchild)

if count(root)>1:
  root.delete(10, root.key)
  print("\nYou want me delete an item. Fine, have it your way you mongrol. The deletion has been done. If you don't believe me, check using traversal function.")
else:
  print("\nI can't perform deletion operation you mongrol. The number of nodes present in the tree is either less than 1 or is completely empty. Huhh!! You always cause me unnecessary trouble.")

root.min_node()
root.max_node()
```
